#**Phi-3**

#The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokonizer separately(although that isn't always necessary)

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load Model and tokonizer

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map = "cuda",
    torch_dtype = "auto",
    trust_remote_code=True,
)

tokonizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

Although we can now use the model and tokenizer directly, It is much easier to warp it in a pipeline object.

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokonizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False,
    )
```

#Finally, We create our prompt as a user and give it to model

```python
messages = [
    {"role":"user", "content": "Create a funny joke about chickens."}
]
# Generate output
output = generator(messages)
print(output[0]['generated_text'])
```
